# We will drop out columns "gender", "postcode", and "age".

```python
# drop columns df1(nan_columns1[2],[3],[4]) from df1
df1 = df1.drop(columns=[nan_columns1[2], nan_columns1[3], nan_columns1[4]])
df1.head(5)
```

```python
# check for every column in df1 nan values exist and if so, print the column name, and store their indexes in a new variable
nan_values1 = df1.isna().any()
nan_columns1 = nan_values1[nan_values1 == True].index
print(nan_columns1)
```

```python
# print uinique values of columns 'is_subscribe' and 'is_sso_user'
print(df1['is_subscriber'].unique())
print(df1['is_sso_user'].unique())
```

```python
df1['is_subscriber'] = df1['is_subscriber'].astype('category')
df1['is_subscriber'] = df1['is_subscriber'].cat.codes
df1['is_sso_user'] = df1['is_sso_user'].astype('category')
df1['is_sso_user'] = df1['is_sso_user'].cat.codes
df1.head(5)
```

```python
# print unique values of device_type column
print(df1['device_type'].unique())
```

```python
# one hot encode the device_type column
df1 = pd.get_dummies(df1, columns=['device_type'])
df1.head(5)
```
